```python
import pandas as pd
import matplotlib.pyplot as plt
```

                                   #BAR PLOT

```python
x = ["Python", "JavaScript", "Java", "C#", "C++", "Ruby", "PHP", "Swift", "Go", "R"]
y = [29.9, 19.1, 15.7, 9.2, 7.8, 4.6, 3.9, 2.8, 2.4, 2.1]
# c = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#9467bd", 
#           "#8c564b", "#e377c2", "#7f7f7f", "#bcbd22", "#17becf"]
c = ["#000000", "#2F4F4F", "#4B0082", "#2E2E2E", "#3C3C3C",
                   "#6E6E6E", "#4A4A4A", "#1C1C1C", "#3D3D3D", "#808080"]
```

```python
plt.bar(x,y , label=x , color=c)
plt.legend()
plt.xlabel('Programming Languages')
plt.ylabel('Popularity (%)')
plt.title("Popular Programming languages" , color = "darkblue")
plt.xticks(rotation=45)
plt.show()
```
