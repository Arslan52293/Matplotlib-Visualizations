```python
import matplotlib.pyplot as plt

# Data
x = [71.0, 111.0]
y = ["PMLN-2013-2018", "P-T-I-2018-2022"]
c = ["lightgreen", "darkgreen"]

# Create the pie chart with enhancements
plt.figure(figsize=(8, 6))
plt.title("Total Public Debt and Liabilities of Pakistan", color="navy", fontsize=16, fontweight='bold')
plt.pie(x, labels=y, colors=c, autopct="%0.2f%%", explode=[0.05, 0.05], shadow=True, startangle=140, textprops={'fontsize': 12})

# Adjust legend position and style
plt.legend(loc="upper left", bbox_to_anchor=(1, 0.8), fontsize=10, frameon=False)

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [71.0, 111.0]
y = ["PMLN-2013-2018", "PTI-2018-2022"]
c = ["lightgreen", "darkgreen"]

# Create the pie chart with enhancements
plt.figure(figsize=(8, 8))  # Set the figure size for a better aspect ratio
plt.title("Total Public Debt and Liabilities of Pakistan", color="navy", fontsize=16, fontweight='bold')  # Enhanced title

# Create the pie chart with additional style
plt.pie(x, labels=y, colors=c, autopct="%0.2f%%", startangle=140, shadow=True, 
        wedgeprops={'edgecolor': 'black', 'linewidth': 1.5},explode=[0.05, 0.05])  # Adding shadows and edge color

# Add a legend with enhanced styling
plt.legend(loc="upper left", fontsize=12, title="Governing Periods", title_fontsize='13', frameon=True)

# Show the chart
plt.show()
```
