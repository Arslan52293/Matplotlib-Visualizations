#BOllywood Actors Ranking

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("BollywoodActorRanking.csv")
df.head(10)
```

```python
df.info()
```

```python
df_subset = df.head(30)
# List of 30 shades of rose color
colors = [
   '#E0F2F1', '#B2DFDB', '#80CBC4', '#4DB6AC', '#26A69A',
    '#009688', '#00897B', '#00796B', '#00695C', '#004D40',
    '#A0D9D1', '#7ACCC4', '#4EB3B0', '#26A6A2', '#008C92',
    '#00796B', '#00695C', '#004D40', '#004D40', '#003B36',
    '#C2C2C2', '#9B9B9B', '#6D6D6D', '#3F3F3F', '#2C2C2C',
    '#00695C', '#004D40', '#003D33', '#003D33', '#00251A',
    '#003B2F'
]
```

```python
plt.figure(figsize=(12, 8))
plt.bar(df_subset["actorName"],df_subset["normalizedRating"],width=0.7,color=colors)
plt.xticks(rotation=90, ha='center', fontsize=10)

plt.xlabel("Actor Name")
plt.ylabel("Normalized Rating")
plt.title("Top 30 Actors by Normalized Rating",color="teal")

plt.subplots_adjust(bottom=0.3)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.bar(df_subset["actorName"],df_subset["movieCount"],width=0.7,color=colors)
plt.xticks(rotation=90, ha='center', fontsize=10)

plt.xlabel("Actor Name")
plt.ylabel("Normalized Rating")
plt.title("Actors with movie Count")

plt.subplots_adjust(bottom=0.3)
plt.show()
```
