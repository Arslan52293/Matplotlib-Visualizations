```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
num = [36, 20, 35, 44, 20, 13, 22, 42, 48, 28, 42, 56, 52, 51, 47, 12, 16,                  #use upper data
       18, 27, 21, 31, 24, 52, 14, 26, 25, 23, 13, 37, 23, 43, 28, 39, 35,             
       46, 46, 59, 34, 58, 22, 15, 13, 21, 30, 25, 46, 13, 19, 15, 27]
```

```python
plt.hist(num, color="green", rwidth=0.8, label="Python")
plt.title("HistogramPlot",color="green")
plt.xlabel("Python - Data")
plt.ylabel("no")
plt.axvline(45,color="red",label="Line")
plt.grid(color="grey")
plt.legend()
plt.show()
```
