                                         #PIE CHART

```python
import matplotlib.pyplot as plt
```

```python
# List of 10 countries with high populations
y = ["China","India","United States","Indonesia","Pakistan"]
x = [30.21,30.65,7.03,5.66,5.20]
c = ["lightblue","pink","yellow","lightgrey","lightgreen"]
ex = [0.1,0.1,0.1,0.1,0.4]
```

```python
plt.pie(x, labels=y, explode=ex, colors = c, autopct="%0.2f%%", shadow=True, radius=1.1, startangle=180)
plt.legend(loc=4)
plt.show()
```

```python
plt.pie(x, labels=y, colors = c, autopct="%0.2f%%", shadow=True, radius=1.1, startangle=180 , 
        counterclock=False)
plt.legend(loc=1)
plt.title("Countries Population",color="blue")
plt.show()
```

              # Double Pie Chart

```python
y = ["China","India","United States","Indonesia","Pakistan"]
x = [30.21,30.65,7.03,5.66,5.20]
x1 = [10,20,30,40,50]
c = ["lightblue","pink","yellow","lightgrey","lightgreen"]
c1 = ["blue","purple","orange","grey","green"]
```

```python
plt.pie(x, labels=y, radius=1.5,colors=c)
plt.pie(x1, radius=1, colors=c1)
plt.show()
```

# Donut Pie Chart

```python
color=["navy","darkgreen","indigo","#800000","#36454F"]
plt.pie(x, labels=y, radius=1.5,colors=color)
plt.pie([1],colors="white")
plt.show()
```
